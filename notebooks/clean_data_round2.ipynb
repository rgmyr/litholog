## Fill `NaN` rows

```python
nan_idxs = beds.isna().any(axis=1)
beds[nan_idxs]
```

```python
# add missing value indicators
missing_tops = beds.gs_tops_mm.apply(np.isnan)

missing_sample = np.logical_and(beds.grain_size_mm.apply(lambda x: np.isnan(x).any()).values,
                                  ~missing_tops.values)

beds['missing_tops'] = missing_tops.astype(float)
beds['missing_sample'] = missing_sample.astype(float)
beds[nan_idxs]
```

```python
from litholog.wentworth import gs2psi

DEFAULT_GS = 0.016 #mm


def snd_shl(gs):
    """
    Threshold `gs` to binary sand/shale indicator.
    """
    return 0 if (gs2psi(gs) <= -4) else 1


default_gs = {
    **{c : DEFAULT_GS for c in beds.columns if 'gs' in c},
    'snd_shl' : 0.0
}


def fill_na_rows(df, nan_idxs, fill_gs=0.016):
    # Need two kinds of fill. 
    for i in list(nan_idxs):
        row = df.loc[i, :]
        
        # One for where gs_tops_mm + all samples are NaN
        if np.isnan(row.gs_tops_mm):
            df.loc[i, default_gs.keys()] = default_gs.values()
            df.at[i, 'depth_m'] = np.array([row.tops, row.bases])
            df.at[i,'grain_size_mm'] = np.array([DEFAULT_GS, DEFAULT_GS])
                                                       
        # One for beds with a single NaN sample.
        else:
            bad_idxs = np.isnan(row.grain_size_mm)
            new_depths, new_values = row.depth_m[~bad_idxs], row.grain_size_mm[~bad_idxs]
                                      
            sample_weights = np.diff(np.abs(np.append(new_depths, row.bases) - row.tops)) 
            new_mean = np.average(new_values, weights=sample_weights)
                                                       
            df.at[i, 'mean_gs_mm'] = new_mean
            df.at[i, 'snd_shl'] = snd_shl(new_mean)                                           
            df.at[i, 'max_gs_mm'] = max(new_values.max(), row.gs_tops_mm)
            df.at[i, 'depth_m'] = new_depths
            df.at[i, 'grain_size_mm'] = new_values

    return df
```

```python
new_df = fill_na_rows(beds, np.where(nan_idxs)[0])
new_df[nan_idxs]
```

## Add `_psi` columns (transform of `_mm` columns)

```python
cols_list = new_df.columns.tolist()

for colname in cols_list:
    if '_mm' in colname:
        psi_vals = new_df[colname].apply(lambda x: np.round(gs2psi(x), 4))
        new_df.insert(new_df.columns.tolist().index(colname) + 1, 
                      colname.replace('_mm', '_psi'), psi_vals)

new_df
```

## Create `rel_depth_m` column (relative depths)

```python
tops = new_df['tops'].values
abs_depths = new_df['depth_m'].values

rel_depths = [np.round(np.abs(tops[i] - abs_depths[i]), decimals=4) for i in range(tops.size)]

new_df = new_df.assign(rel_depth_m=rel_depths)
new_df
```

## Create seperate `snd` and `shl` columns

```python
        # split sand/shale into seperate indicators
new_df['snd'] = new_df.snd_shl
new_df['shl'] = (-1)*(new_df.snd - 1)
new_df
```
