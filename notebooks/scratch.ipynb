```python
def safelen(x):
    """
    Return the length of an array or list, or 1 for numeric literals.

    **Should we throw an error for arrays with dim > 1?
    """
    if isinstance(x, (np.ndarray, np.generic)):
        _x = np.squeeze(x) if x.ndim > 1 else x
        _shape, _ndim = _x.shape, _x.ndim
        assert _ndim <= 1, 'Using `safelen` of array with ndim > 1 is not allowed'
        return _shape[0]
    else:
        try:
            return len(x)
        except TypeError:
            return 1
```

```python
safelen(np.array([1]))
```

```python
safelen(np.array([[1, 2, 3]]))
```

```python
safelen([1, 2, 3])
```
